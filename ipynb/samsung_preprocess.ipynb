```python
# 키움 OPEN API로 추출한 삼성 일별 주가
import pandas as pd
ss_df = pd.read_csv('./samsung_data.csv')
ss_df = pd.DataFrame(ss_df)
ss_df
```

```python
del ss_df['Unnamed: 0']
ss_df.reset_index(drop=True,inplace= True)
```

```python
ss_df.drop(['시가', '고가', '저가', '전일비','거래량','금액(백만)','신용비','개인','기관','외인수량','외국계','프로그램','외인비','체결강도','외인보유','외인비중','신용잔고율'], axis=1, inplace=True) 
```

```python
ss_df.dtypes
```

```python
ss_df
```

```python
# 결측치 제거
# 날짜데이터(float64) -> datetime
# 종가(float64) -> int64로 변경 후 절대값
# 외인순매수 기관순매수	개인순매수 (object) -> "--"값을 "-"로 변경 후 int64로
```

```python
ss_df = ss_df.dropna(axis=0,how='all',subset= None)
ss_df
```

```python
#import datetime

#ss_df["날짜"] = pd.to_datetime(ss_df["날짜"])

ss_df["날짜"]=ss_df["날짜"].astype("int64")
```

```python
#절대값 변환 함수
def _abs(list):
    new_list = []
    for i in range(len(list)):
        list[i]= abs(list[i])
        new_list.append(list[i])
    return new_list

#int64로 변환 
ss_df["종가"].astype("int64")

ss_df["종가"]=_abs(ss_df["종가"])
```

```python
#object 일때 문자열 replace하는 함수

def del_minus(list):
    new_list = []
    for i in range(len(list)):
        list[i]=list[i].replace('--','-')
        list[i]=list[i].replace('+-','-')
        new_list.append(list[i])
    return new_list


ss_df["외인순매수"]=del_minus(ss_df["외인순매수"])
ss_df["기관순매수"]=del_minus(ss_df["기관순매수"])
ss_df["개인순매수"]=del_minus(ss_df["개인순매수"])

# 후에 int64로 변환
ss_df[["외인순매수","기관순매수","개인순매수"]]=ss_df[["외인순매수","기관순매수","개인순매수"]].astype("int64")
```
