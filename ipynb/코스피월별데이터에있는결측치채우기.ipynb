```python
import pandas as pd
```

```python
fill=pd.read_csv('C:/Users/jangd/OneDrive/사진/문서/GitHub/git_hub_0420/python/project/dataset/pre_kospi.csv')
fill.drop('등락율',axis=1,inplace=True)
fill['날짜']=pd.to_datetime(fill['날짜'])
fill.dtypes
```

```python
fill_first=[]

for i in range(0,6322):
    fill_first.append(i)
fill_first=pd.DataFrame(fill_first)
fill_first['날짜']=pd.date_range(start='2005-01-03',end='2022-04-25')
fill_first.columns=['처음','날짜']
fill_first.drop('처음',axis=1,inplace=True)
fill_first['날짜']=pd.to_datetime(fill_first['날짜'])
fill_first.dtypes
```

```python
fill_merge=pd.merge(left = fill_first, right = fill, how = "left", on = "날짜")
fill_merge
```

```python
fill_full=fill_merge.bfill()
fill_full.dtypes
```

```python
import datetime

def last_day_of_month(any_day):
    next_month = any_day.replace(day=28) + datetime.timedelta(days=4)  # this will never fail
    return next_month - datetime.timedelta(days=next_month.day)
month_last_day=[]
for j in range(2005,2023):
    for month in range(1, 13):
        month_last_day.append(last_day_of_month(datetime.date(j, month, 1)))
month_last_day_df=pd.DataFrame(month_last_day)
month_last_day_df.columns=['날짜']
month_last_day_df['날짜']=pd.to_datetime(month_last_day_df['날짜'])
month_last_day_df.dtypes
```

```python
fill_final=pd.merge(left = month_last_day_df, right = fill_full, how = "left", on = "날짜")

fill_final.drop(index=[207,208,209,210,211,212,213,214,215],inplace=True)
fill_final.columns=['날짜','아','종가','개인순매수','외국인순매수','기관순매수']
fill_final.drop('아',axis=1,inplace=True)
```

```python
fill_final["전월대비 등락률"] = (fill_final["종가"]-fill_final["종가"].shift(1))*100/fill_final["종가"].shift(1)
```

```python
fill_final.drop(['종가','개인순매수','외국인순매수','기관순매수'],axis=1,inplace=True)
```

```python
fill_final.to_csv('결측치_채워준_월별데이터.csv')
```
