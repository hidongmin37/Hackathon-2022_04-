```python
import requests as re
from bs4 import BeautifulSoup
from pandas import DataFrame, Series
import numpy as np
import pandas as pd

#본인 PC에 맞는 폴더 위치를 입력하시면 됩니다.
folder_adress = '/content/drive/MyDrive/Untitled Folder'


def get_trading_trend():
    url = 'http://finance.naver.com/sise/investorDealTrendDay.nhn?bizdate=20220425&sosok=&page='

    date_list = []
    private = []
    foreign = []
    institution = []

    for i in range(1,500):
        url_ = re.get(url + str(i))
        url_ = url_.content
        html = BeautifulSoup(url_,'html.parser')
    
        body = html.find('body')
    
        tr = body.find_all('tr')

        for r in tr:
            date = r.find('td',{'class':'date2'})
    
            if date != None:
                date = date.text.strip().replace('.','-')
                
                
                # 날짜가 중첩되지 않으면 계속 크롤링 : 
                if not date in date_list:
                    date_list.append(date)
                    print(date)
                
                # 더 이상 자료가 없어, 날짜가 중첩되면 : 크롤링 완료
                else:
                    Data = DataFrame(index = date_list)
                    Data['개인'] = private
                    Data['외국인'] = foreign
                    Data['기관'] = institution
                    
                    Data.to_csv(folder_adress+'/trading_trend.csv')
                    return Data
                
                td = r.find_all('td')
        
                count = 0
        
                # 앞에서 3개 값 '개인' , '외국인' , '기관' 만 가져온다
                for d in td:
                    if count != 3:
                        d = d.text.replace(',','')
                    try:
                        d = int(d)
                    
                        if count == 0 :
                            private.append(d)
                        elif count == 1 :
                            foreign.append(d)
                        else:
                            institution.append(d)
                        
                        count += 1
                    
                    except:
                        count = count
    

    
    return Data
            
Data = get_trading_trend()
```
