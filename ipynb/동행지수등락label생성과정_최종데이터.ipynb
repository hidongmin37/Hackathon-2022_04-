```python
import pandas as pd
import numpy as np
```

```python
final=pd.read_csv('./data/더미변수없는최종데이터206행.csv')

final['동행지수']
final=pd.DataFrame(final)
final.set_index('날짜',inplace=True)
final
```

```python
final.info()
```

```python
final
```

```python
# diff를 통해 차분시행 후, 데이터 Nan 값 제거

kospi_diff = final.diff(1)['동행지수']
kospi_diff = pd.DataFrame(kospi_diff)
print('diff 후 \n', kospi_diff.head())

print('-'*10)
# shift를 통해 1/4일 데이터를 1/5일로 미룸

kospi_shift = final['동행지수'].shift(1)
kospi_shift = pd.DataFrame(kospi_shift)
print('shift 후 \n', kospi_shift.head())
```

```python
final['동행지수등락률']= (kospi_diff / kospi_shift) * 100
#final.drop('등락률',axis=1,inplace=True)
final.sort_index(ascending=False)

final
```

```python
#NAN값 채우기
final.loc['2005-01-31','동행지수등락률']=-0.1
```

```python
final['동행등락label']=np.arange(206)
for i in range(206):
    if final['동행지수등락률'][i]>=0:
        final['동행등락label'][i]=1
    else:
        final['동행등락label'][i]=0
```

```python
final.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
final.to_csv('end.csv')
```
