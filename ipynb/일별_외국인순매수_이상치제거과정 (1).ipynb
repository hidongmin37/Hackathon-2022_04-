### 외국인 일별 순매수량 이상치 제거 
 - 외국인 순매수 월별합 정규성 검정을 했을때 정규성에 맞지 않음
 - 그래서 먼저 일별 이상치를 제거하기로 함
 - 그 후 일별 이상치를 제거한 데이터로 월별로 합산함
 -> 정규성 만족함

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy as sp
```

```python
df = pd.read_csv('./data/pre_kospi.csv')
df
```

```python
df['외국인순매수'].describe()
```

### 상자수염그림(boxplot) 그래프
-> 이상치가 많이 나옴을 볼 수 있음

```python
# box-plot 그래프 
plt.figure(figsize = (5,12))
sns.boxplot(y='외국인순매수', data = df)
plt.title('vertical graph', size = 15)
plt.show()
```

### 이상치 제거 함수 get_outlier 만들기

```python
import numpy as np
#이상치 제거 함수 get_outlier 만들기

def get_outlier(df=None, column=None, weight=1.5):
  # target 값과 상관관계가 높은 열을 우선적으로 진행
  quantile_25 = np.percentile(df['외국인순매수'].values, 25)
  quantile_75 = np.percentile(df['외국인순매수'].values, 75)

  IQR = quantile_75 - quantile_25
  IQR_weight = IQR*weight
  
  lowest = quantile_25 - IQR_weight
  highest = quantile_75 + IQR_weight
  
  outlier_idx = df['외국인순매수'][ (df['외국인순매수'] < lowest) | (df['외국인순매수'] > highest) ].index
  return outlier_idx
```

```python
df.head(3)
```

```python
# 함수 사용해서 이상치 값 삭제
outlier_idx = get_outlier(df=df, column='외국인순매수', weight=1.5)
df['외국인순매수'].loc[outlier_idx]
```

### 이상치 있는 날만 뽑아냄 221개 이상치 (4281개 중 221개 = 약 5%)

```python
df_out = df.loc[outlier_idx]
df_out
```

```python
# 이상치 제거한 데이터 프레임
df_outt = df_out.drop('Unnamed: 0',axis=1)
```

```python
#날짜로 인덱스 추가
df_outt.set_index('날짜', inplace=True)
df_outt
```

### 이상치가 발생한 일자 뽑기

```python
df_out_date = df_out['날짜']
df_out_date
df_out_date.to_csv('./data/일별_외국인매수량_이상치발생날짜.csv')
```
