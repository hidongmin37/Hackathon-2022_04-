```python
# kospi_df 와 df 병합

def _combine(kospi_df,df):
    k_df_list = kospi_df["날짜"]
    df_list = df["날짜"]
    for i in range(len(kospi_df)):
        if k_df_list[i] == df_list[i]:
            kospi_df.drop(['날짜'], axis=1, inplace=True) 
            all_df = pd.concat([kospi_df,df], axis = 1)
        return all_df

complete_df=_combine(kospi_df,df)
```
