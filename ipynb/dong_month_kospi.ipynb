```python
df_1=pd.read_csv('C:/Users/jangd/OneDrive/사진/문서/GitHub/git_hub_0420/python/project/dataset/kospi.csv')
df_1.drop(['등락율'],axis=1,inplace=True)
df_1
```

```python
# diff를 통해 차분시행 후, 데이터 Nan 값 제거

kospi_diff = df_1.diff(1)['종가'].dropna()
kospi_diff = pd.DataFrame(kospi_diff)
print('diff 후 \n', kospi_diff.head())

print('-'*10)
# shift를 통해 1/4일 데이터를 1/5일로 미룸

kospi_shift = df_1['종가'].shift(1).dropna()
kospi_shift = pd.DataFrame(kospi_shift)
print('shift 후 \n', kospi_shift.head())
```

```python
df_1['등락률']= (kospi_diff / kospi_shift) * 100
df_1.sort_index(ascending=False)
# df_1['Day_Name'] = df_1['Date'].dt.day_name()
df_1.drop('date',axis=1,inplace=True)
df_1
```
